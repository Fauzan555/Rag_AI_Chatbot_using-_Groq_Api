#  Adding Chat Memory (Context Awareness)

```python
from langchain_groq import ChatGroq
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.messages import HumanMessage
import os

# Set API Key
os.environ["GROQ_API_KEY"] = "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"

# Initialize Groq model
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile"
)

# Create memory
history = InMemoryChatMessageHistory()

# First message
history.add_message(
    HumanMessage(content="Hi, my name is Alex")
)

response1 = llm.invoke(history.messages)
print(response1.content)

# Save AI response to memory
history.add_message(response1)

# Second message
history.add_message(
    HumanMessage(content="What is my name?")
)

response2 = llm.invoke(history.messages)
print(response2.content)
```

# Creating Embeddings and Vector Store

```python
from langchain_community.document_loaders import TextLoader
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Load document
loader = TextLoader(
    r"c:\PROJECTS\AI Chatbot\data.txt.txt",
    encoding="utf-8"
)

documents = loader.load()

# Embedding model
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Create vector database
vectorstore = FAISS.from_documents(
    documents,
    embedding=embeddings
)

print("FAISS vector store created successfully")
```

# Building a Retrieval Chain

```python
from langchain_groq import ChatGroq
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
import os

# Set Groq API Key
os.environ["GROQ_API_KEY"] = "your_groq_api_key"

# Initialize Groq LLM
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile"
)

# Memory
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)

# Conversational Retrieval Chain
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    memory=memory
)

print("QA Chain Created Successfully")
```

# Ask Questions From Your Data

```python
response = qa_chain.invoke(
    {"question": "What is the article about?"}
)

print(response["answer"])
```
